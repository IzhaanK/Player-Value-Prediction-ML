## Checking for null values

```python
# Creating a list of columns in which more than 50 % data is null / NaN 
null_columns = []
for i in data.columns:
    if (data[i].isnull().sum() / len(data)*100) > (50) :
        null_columns.append(i)
print('More than 50% null data Columns are : ','\n',null_columns)
```

there are no columns with missing values (thanks to this dataset)

```python
# Dropping unwanted columns
data.drop(['Known As','Image Link','National Team Image Link'],axis=1,inplace=True)
```

```python
# Changing data type of date columns
data.rename(columns={'Joined On': 'joined'}, inplace=True)
data.rename(columns={'Wage(in Euro)': 'value_eur'}, inplace=True)
data['Release Clause'] = pd.to_datetime(data['Release Clause'])
data['joined'] = pd.to_datetime(data['joined'])
```

## Exploratory Data Analysis

```python
# Separating Categorical and Numerical columns
cat_columns = []                                        # create a categorical empty list
num_columns = []                                        # create a Numerical empty list
for i in data.keys() :                                  # use a for loop to iterate over the data set
    if data[i].dtype == 'O' :                           # give condition for Objective Data Type
        cat_columns.append(i)                           # append objective columns
    else :
        num_columns.append(i)                            # append Non-objective columns
        
print('Categorical Columns : \n',cat_columns,'\n')

print("Numerical Columns : \n",num_columns)
```

```python
# The data has been segregated in lists now we will create a dataframe

num_data = data[num_columns]   # Creating a data Frame of Numerical features
cat_data = data[cat_columns]   # Creating a data Frame of Categorical features
```

## EDA for Numerical Columns

```python
# Univariate Analysis of first 40 Numerical features

plt.figure(figsize=(30,90))                   # Set Canvas size
plotnumber = 1                                # Create a variable
for i in num_data.keys():                     # use for loop to iterate Numerical Variable
    plt.subplot(20,4,plotnumber)              # set number of rows & columns according to No.of Variable
    sns.histplot(x=data[i],color="green")   # plot Histogram
    plt.xticks(rotation=90,fontsize=20)
    plt.xlabel(i,fontsize=25)
    plt.yticks(fontsize=20)
    plotnumber = plotnumber + 1
plt.tight_layout()                            # to avoid overlapping of plots
```

## EDA for Categorical Features


```python
# Counting the number of players per nationality
nationality_count = data['Nationality'].value_counts()

# Focusing only on the top 10 nationalities(rest of the countries will be summed up as others)
top_nations = nationality_count.head(10)  # Get top 10 nationalities

# Creating a bar plot for the top 10 nations
plt.figure(figsize=(10, 6))
sns.barplot(x=top_nations.values, y=top_nations.index, palette="viridis")

# Adding labels and title
plt.xlabel('Number of Players')
plt.ylabel('Country')
plt.title('Top 10 Countries with Most Players')

# Showing the plot
plt.show()
```

```python
# Grouping by nationality and calculate player count and total value per country
country_stats = data.groupby('Nationality').agg(
    player_count=('Full Name', 'count'),
    total_value=('Value(in Euro)', 'sum')
).sort_values(by='player_count', ascending=False)

# Getting the top 5 countries by player count
top_5_countries = country_stats.head(5)

# Calculating percentage of total players and percentage of total value they contribute
total_players = country_stats['player_count'].sum()
total_value = country_stats['total_value'].sum()

top_5_countries['player_percentage'] = (top_5_countries['player_count'] / total_players) * 100
top_5_countries['value_percentage'] = (top_5_countries['total_value'] / total_value) * 100

# Displaying the results
print(top_5_countries)
```

## Insights
-Brazil contributes 10% of all players but 15% of the total value.

-Germany contributes 8% of players but only 6% of the total value.

-Argentina contributes 7% of players but 12% of the total value.

Key takeaway: This insight reveals that while certain countries (like Brazil or Argentina) may not produce the most players, their players tend to have a higher value on average compared to countries that produce more players. Conversely, countries like Germany may produce a high number of players, but their players' market value is not as high.

This provides a clear view of the countries that dominate both in terms of numbers and value and those where the quality (value) surpasses the quantity.








```python
# Calculating average player value for each age group 
age_value_group = data.groupby('Age')['value_eur'].mean().reset_index()

# Plotting the age vs. average market value
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='value_eur', data=age_value_group)
plt.title('Average Market Value by Age')
plt.xlabel('Age')
plt.ylabel('Average Market Value (in EUR)')
plt.show()

# Insight: Finding the age range where players hit their peak value
peak_value_age = age_value_group.loc[age_value_group['value_eur'].idxmax()]['Age']
print(f"Players reach their peak market value at age {peak_value_age}.")
```

```python
# Grouping by 'best_position' and calculating total value
position_value_group = data.groupby('Best Position')['value_eur'].sum().reset_index()
position_value_group = position_value_group.sort_values(by='value_eur', ascending=False)

# Calculating total value and top 3 positions' contribution
total_value = position_value_group['value_eur'].sum()
top_3_positions = position_value_group.head(3)
top_3_value = top_3_positions['value_eur'].sum()
top_3_contribution_percentage = (top_3_value / total_value) * 100


# Plotting the value contribution of each position
plt.figure(figsize=(10, 6))
sns.barplot(x='Best Position', y='value_eur', data=position_value_group.head(10))
plt.title('Top 10 Positions by Total Market Value')
plt.xlabel('Position')
plt.ylabel('Total Market Value (in EUR)')
plt.show()


print(f"The top 3 positions (ST, CAM, RW/LW) contribute {top_3_contribution_percentage:.2f}% of the total player market value.")
```

Only 3 positions—Striker (ST), Attacking Midfielder (CAM), and Winger (RW/LW)—account for 45.69% of the total player market value, despite representing fewer player numbers.

```python
# Grouping by 'preferred foot' and calculating average value
foot_value_group = data.groupby('Preferred Foot')['value_eur'].mean().reset_index()

# Calculating the percentage difference between left and right foot players
left_foot_value = foot_value_group[foot_value_group['Preferred Foot'] == 'Left']['value_eur'].values[0]
right_foot_value = foot_value_group[foot_value_group['Preferred Foot'] == 'Right']['value_eur'].values[0]
value_difference_percentage = ((left_foot_value - right_foot_value) / right_foot_value) * 100



# Plotting preferred foot vs. average market value
plt.figure(figsize=(8, 6))
sns.barplot(x='Preferred Foot', y='value_eur', data=foot_value_group)
plt.title('Average Market Value by Preferred Foot')
plt.xlabel('Preferred Foot')
plt.ylabel('Average Market Value (in EUR)')
plt.show()
print(f"Left-footed players have a {value_difference_percentage:.2f}% higher market value than right-footed players.")
```
